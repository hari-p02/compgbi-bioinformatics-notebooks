The string is enclosed in double quotes if the string contains a single quote and no double quotes, otherwise it can be enclosed in single quotes.

# **Functions**

A function in Python is a group of related statements that perform a particular task. Defining a function involves giving a name to a set of actions you want the computer to execute. All functions must be called in order to be executed. A call to a function means you tell the computer to execute those sets of actions.

A function is defined using the ***def*** keyword.

```python
def new_function():
  print("I have created my first function!")

new_function() #calling the function to execute
```

Parameters are the information that is passed into functions. It is the variable listed within the parenthesis of the function definition. This is sent to the function when called. These parameters will be used as variables in the body of the function.

```python
def name(firstname):
  print(firstname + " Smith")

name("Jane")
name("John")
name("Jerry")
```

You can pass multiple parameters to a function by listing the variables in the function definition. If the function has 2 parameters, then the function must be called with 2 parameters

```python
def name(firstname, lastname):
  print(firstname + lastname)

name("Jane", " Smith")
name("John", " Doe")
name("Jerry", " Green")
name("Jen")
```

# **Object-Oriented Programming**

***Classes, Objects, & Methods...***



Python is Object-Oriented Programming (OOP),  which is a programming model that is based on the concept of classes and objects. Classes are a code template for creating objects, and these classes represent broad categories that share attributes.

> Class attributes are variables of a class that represent the different features of the class.


> Classes contain functions, which are called methods. These methods pertain only to the objects of the class's type, and they help to perform an action particular to the type of object.


> An object is an instance of a class. Many objects can be created under the same class type. Every object can have unique attributes that are defined within the class.

*Let's see an example...*

```python
class Person:
  def __init__(self, firstname, lastname):
    self.firstname = firstname
    self.lastname = lastname

name = Person("Jan", "Smith")

print(name.firstname)
print(name.lastname)
```

In the example above, we created a class and initialized the object.

Some classes in Python have the __init__() function, which executes when the class is initialized. This function assigns values to object properties.

The self parameter references the current object of the class and is used to access variables within the class.

In this next example, we will create a method to the class Person.

```python
class Person:
  def __init__(self, firstname, lastname, age):
    self.firstname = firstname
    self.lastname = lastname
    self.age = age

  def age(self, age):
    self.age = age
    print("My name is " + self.firstname + " and I am " + self.age + " years old.")

name = Person("Jan", "Smith", "18")
print(name.firstname)
print(name.lastname)
print(name.age)

```

In the example above, calling a method requires the **.** operation. We apply the method *age* to the variable *name* in order to use it.

# **Control Flow**

In programming, a computer runs through code in order line by line. This order of executing code is called control flow. Structures that change the control flow are conditional statements, like if statements, and loops, like for and while loops. Thus, these structures are just decisions that a computer makes.

There are 3 basic types of control flow structures:


1.   Sequential -- statements executed in order.
2.   Selection -- choosing between 2 or more different paths (if statements).
3. Repetition -- repeating a chunk of code multiple times in a row (looping).

http://net-informations.com/python/flow/default.htm

  

> # **If Statements**



If statements are used for decision-making operations. This contains a block of code that runs only when the condition in the if statement is true. If the condition is false, then the code either does not run, or an optional else statement is executed which contains code for the else condition.

```python
value = 0 #try changing this value to -3 or 0.

if value >=0 :
  print("The number is 0 or a positive number.")
else:
  print("The number is a negative number.")
```

You can also check for multiple expressions by using the ***elif*** statement, which is short for else if.

```python
value = 2

if value == 0:
  print("The number is 0.")
elif value > 0:
  print("The number is positive.")
else:
  print("The number is negative.")
```



> # **For Loop**



A loop is a sequence of instructions that repeats until a certain condition is reached. The for loop in Python is used to loop, or iterate, over an iterable object. Objects like strings, lists, etc. are iterable.

```python
letters = ['a', 'b', 'c', 'd', 'e']

for letter in letters: #iterate over the list
  print(letter)
```

```python
numbers = [1, 2, 3, 4, 5]

product = 1 #variable to store the product

for num in numbers:
  product = product * num

print("The product is", product)
```

As you can see in the code above, the for loop is performing the operation 1 * 2 * 3 * 4 * 5, which is equal to 120.

To iterate over a sequence of numbers, use the built-in function ***range()***. This function will output each value, starting from 0 all the way up to (but not including) the chosen number.

```python
for i in range(10):
  print(i)
```

```python
for i in range(1, 10):
  print(i)
```

> # **While Loop**


The while loop is used to loop over a chunk of code so long as the condition evaluates to true.

```python
i = 0
while i < 10:
  print(i)
  i += 1 #increment the value of i, otherwise the loop will continue forever
```

We can use the ***break*** statement to stop the loop from continuing even if the condition in the while loop is true.

```python
i = 0
while i < 10:
  print(i)
  break #does not iterate over the other values, just stops at the initial value of 0
  i +=1
```

We can use the ***continue*** statement to stop the current iteration and move onto the next iteration.

```python
i = 0
while i < 10:
  i+= 1
  if i == 2: continue #moves on from 2 and continues with 3
  print(i)
```

We can also use ***else*** statements in while loops to run a sequence of code when the condition of the while loop is not true anymore.

```python
i = 0
while i < 10:
  print(i)
  i += 1
else:
  print("The value i is not less than 10 anymore.")
```

# **Modules**

A module in Python is a file of Python code. These modules can perform a specific task, and you can have multiple modules make up a greater objective for a larger application.


A module is referenced by the name of its file without the .py extension. Thus, a file called ***calculations.py*** has the module name ***calculations***.

```python
# calculations.py

def addition(x, y):
  return x + y

def multiplication(x, y):
  return x * y
```

The calculations module has two functions that add two numbers, x and y, and multiply two numbers.  



> # **Importing Modules**



You can gain access to code in another module by importing it. The ***import*** statement is the most common way to do this.

```python
import module_name
```

For example, to use the calculations module in the morecalculations.py file, use the statement below:

```python
# morecalculations.py
import calculations
```

Importing a module makes all of the code in that file execute in the file where you are importing. In the example above, all of the code in calculations.py will be executed in the file morecalculations.py . Thus, you can access functions, variables, and other objects from the imported module in the current module.

The following code is the syntax for calling a function from the imported module in the current module:

```python
module_name.function_name()
```

Here is an example of how to use the **addition()** and **multiplication()** functions defined in the ***calculations*** module in the ***morecalculations.py*** file.

```python
# morecalculations.py
import calculations

sum = calculations.addition(
    x = 10,
    y = 20
)
print(sum)

product = calculations.multiplication(
    x = 5,
    y = 2
)
print(product)
```

Output of the above code:

```python
30
10
```

There are numerous ways to import objects defined in another file in your current module, the ***import*** statement is only one of them. Find more on how to import modules [here](https://docs.python.org/3/tutorial/modules.html).

The following is an example of how to use ***matplotlib*** to create a histogram.

```python
from matplotlib import pyplot as plt
import numpy as np

# Creating dataset
a = np.array([22, 87, 5, 43, 56,
			73, 55, 54, 11,
			20, 51, 5, 79, 31,
			27])

# Creating histogram
fig, ax = plt.subplots(figsize =(4, 3))
ax.hist(a, bins = [0, 20, 40, 60, 80, 100])

# Show plot
plt.show()
```

You can use matplotlib to customize your histogram. Using the matplotlib.pyplot.hist() function, there are many different attributes that can help modify the plot to your liking. You can read more about this here: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html.

The following code is an example of how to create a pie chart.

```python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np


# Creating dataset
cars = ['NARUTO', 'JJK', 'AOT',
		'DEMON SLAYER', 'ONE PIECE', 'SPY FAMILY']

data = [23, 17, 35, 29, 12, 41]

# Creating plot
fig = plt.figure(figsize =(5, 3))
plt.pie(data, labels = cars)

# show plot
plt.show()
```

Another example of how to use matplotlib to create different plots is shown in the code below. This code provides three different graphs in the forms of square root, square, and exponential graphs, which are important quantitative measures in bioinformatics.

```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import math

# Placing the plots in the plane
plot1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
plot2 = plt.subplot2grid((3, 3), (0, 2), rowspan=3, colspan=2)
plot3 = plt.subplot2grid((3, 3), (1, 0), rowspan=2)

# Using Numpy to create an array x
x = np.arange(1, 10)

# Plot for square root
plot2.plot(x, x**0.5)
plot2.set_title('Square Root')

# Plot for exponent
plot1.plot(x, np.exp(x))
plot1.set_title('Exponent')

# Plot for Square
plot3.plot(x, x*x)
plot3.set_title('Square')

# Packing all the plots and displaying them
plt.tight_layout()
plt.show()
```
